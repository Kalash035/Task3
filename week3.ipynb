```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Sales': [150, 200, 250, 210, 350, 300],
    'Expenses': [80, 90, 100, 110, 120, 130]
}

# Create DataFrame
df = pd.DataFrame(data)

# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Sales'], color='blue', label='Sales')
plt.bar(df['Month'], df['Expenses'], color='red', bottom=df['Sales'], label='Expenses')

# Customizing the bar chart
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the bar chart
plt.show()
```

```python
# Line Chart
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Sales'], marker='o', color='blue', label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='o', color='red', label='Expenses')

# Customizing the line chart
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Sales and Expenses')
plt.legend()
plt.xticks(rotation=45)
plt.grid()

# Display the line chart
plt.show()
```
